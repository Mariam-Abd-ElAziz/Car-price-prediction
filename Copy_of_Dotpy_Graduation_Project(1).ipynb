```python
data=data[(data['price']<=2e6)]
```

```python
def fill_outliers(data, col):
  Q1 = data[col].quantile(0.25)
  Q3 = data[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  col_avg=data[col].mean()
  outliers = (data[col] < lower_bound) | (data[col] > upper_bound)
  data.loc[outliers, col] = col_avg
  return data
```

```python
data.shape
```

```python
filter_data=['back_legroom','highway_fuel_economy','front_legroom','daysonmarket','horsepower','mileage','wheelbase','maximum_seating']
```

```python
for col in filter_data:
  data = fill_outliers(data, col)
```

```python
categorical_columns = ['body_type', 'engine_type', 'franchise_make']
```

```python
for col in categorical_columns:
  value_counts = data[col].value_counts()
  # Find the threshold for outlier values
  threshold = value_counts.quantile(0.25)

  # Filter the data to remove values with frequency below the threshold
  data = data[data[col].isin(value_counts[value_counts >= threshold].index)]
```

##Evaluate

```python
from sklearn.model_selection import KFold
kf =KFold(n_splits=5)
```

```python
scores=[]
```

```python
for train_index, test_index in kf.split(features,label):
    # Split data into training and testing sets
    x_train,x_test=features[train_index],features[test_index]
    y_train, y_test = label.iloc[train_index],label.iloc[test_index]
    model2.fit(x_train,y_train)
    score = model2.score(x_test, y_test)
    scores.append(score)
```
